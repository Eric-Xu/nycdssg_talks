### Import libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import peakutils
from peakutils.plot import plot as pplot

%matplotlib inline
plt.style.use('ggplot')
```

```python
x = np.linspace(0, 10, 11)
y = np.array([0, 1, 0, 2, 0, 3, 0, 2, 0, 1, 0])

indexes = peakutils.peak.indexes(y, min_dist=2)

plt.figure(figsize=(8,4))
plt.title('min_dist=2')
plt.ylim([np.min(y)-0.2, np.max(y)+0.2])
pplot(x, y, indexes)
```

```python
x = np.linspace(0, 1, 200)
y = np.sin(2*np.pi*5*x) + np.random.randn(200)/5

indexes = peakutils.peak.indexes(y, min_dist=20)

plt.figure(figsize=(8,4))
plt.title('min_dist=20')
plt.ylim([np.min(y)-0.2, np.max(y)+0.2])
pplot(x, y, indexes)
```

```python
x = np.linspace(0, 6, 7)
y = np.array([0, 0.9, 0, 0.6, 0, 0.3, 0])

# Normalized threshold (float between [0., 1.]) 
# Only the peaks with amplitude higher than the threshold will be detected.
indexes = peakutils.peak.indexes(y, thres=0.5)

plt.figure(figsize=(8,4))
plt.title('thres=0.5')
plt.ylim([np.min(y)-0.2, np.max(y)+0.2])
pplot(x, y, indexes)
```
